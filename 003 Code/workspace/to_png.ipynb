```python
from pathlib import Path
from PIL import Image

directory = Path("/workspace/jke_capston/client/public/images")

for jpg_file in list(directory.glob("*.jpg")) + list(directory.glob("*.jpeg")):
    img = Image.open(jpg_file)
    png_path = jpg_file.with_suffix('.png')
    img.save(png_path, 'PNG')
    jpg_file.unlink()
    print(f"{jpg_file.name} -> {png_path.name}")
```
